```python
import pandas as pd
import numpy as np
df=pd.read_csv('aboutus.csv')

```

```python
#Ordinal ENcoding

df['Specialization_encoded'] = df['Specialization'].map({'Data Science & Analytics':1, 'Software Development':2, 'Finance & Investment':3, 'Design & Creative Media':4})
df
```

```python
#Onehot Encoding
spz=df['Specialization']
unq=[]
for x in spz:
  if x in unq:
    pass
  else:
    unq.append(x)
print(unq)
for sp in unq:
    df[sp] = df['Specialization'].map(lambda x: 1 if x == sp else 0)
df
```

```python
imp_col = [col for col in df.columns if df[col].isnull().any()]
print(imp_col)
```

```python
#imputation

total=0
ct=0
for x in df['Age ']:
  if x!=x:
    pass
  else:
    total=total+x
    ct=ct+1
mean_age=total/ct
print(mean_age)

df['Age '] = df['Age '].fillna(mean_age)
df
```
